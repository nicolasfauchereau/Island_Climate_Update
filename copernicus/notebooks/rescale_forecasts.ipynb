```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
data = pd.read_csv('../outputs/distribution_New_Caledonia_v2_MAM_2019.csv')
```

```python
data = data.loc[:,['tpara_masked']]
```

```python
data.hist()
```

```python
data_std = data / data.std()
```

```python
f, axes = plt.subplots(ncols=2, figsize=(12, 6))
sns.distplot(data, rug=True, ax=axes[0])
sns.distplot(data_std, rug=True, ax=axes[1])
axes[0].grid(ls=':')
axes[1].grid(ls=':')
```
